```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as cte
import pandas as pd
```

```python
Datos = pd.read_csv('Tobera_MP_V2.csv')
alpha = Datos['Alpha'].item()
beta = Datos['Beta'].item()
Po = Datos['Presión promedio'].item()
AeAt = Datos['Relación de expansión'].item()
AcAt = Datos['Ac/At'].item()
De = Datos['De'].item()
Dt = Datos['Dt'].item()
Dc = Datos['Dc'].item()

print(De, Dc, Dt)
```

```python
hc = (Dc-Dt)/2
he = (De-Dt)/2

Le = he/np.tan(alpha)
Lc = hc/np.tan(beta)
rond = 2 # Ancho de las rondanas

L_noz = Le + Lc + 2*rond

print(hc, he, L_noz)
print(Le, Lc)
```

```python
import numpy as np
import matplotlib.pyplot as plt



# Convertir ángulos de grados a radianes para cálculos

# Cálculos de coordenadas, suponiendo una simplificación del perfil
# Coordenadas de la entrada
x_c = 0
y_c = Dc/2

# Coordenadas del cuello
x_t = Lc
y_t = Dt/2

# Coordenadas de la salida
x_e = Le + x_t
y_e = De/2

# Crear las líneas del perfil de la tobera
x = [x_c, x_t, x_e, x_e, x_t, x_c]
y = [y_c, y_t, y_e, -y_e, -y_t, -y_c]

# Graficar el perfil de la tobera
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Perfil de la Tobera')
plt.scatter(x, y, color='red')  # Puntos de control

# Añadir líneas acotadas para diámetros
plt.plot([x_c, x_c], [-Dc / 2, Dc / 2], 'k--')
plt.plot([x_t, x_t], [-Dt / 2, Dt / 2], 'k--')
plt.plot([x_e, x_e], [-De / 2, De / 2], 'k--')

# Etiquetas de los diámetros en las líneas
plt.text(x_c-1, 0, f' Dc = {Dc:.2f}', verticalalignment='bottom', horizontalalignment='right', color='black')
plt.text(x_t-1, 0, f' Dt = {Dt:.2f}', verticalalignment='bottom', horizontalalignment='right', color='black')
plt.text(x_e-1, 0, f' De = {De:.2f}', verticalalignment='bottom', horizontalalignment='right', color='black')


plt.title('Perfil de la Tobera')
plt.xlabel('X, mm')
plt.ylabel('Y, mm')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
